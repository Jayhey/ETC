```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
```

```python
url = 'http://www.easylaw.go.kr/CSP/CsmSortRetrieveLst.laf?sortType=gnd&gndIdx='
root_url = 'http://www.easylaw.go.kr/CSP/OnhunqueansLstPopRetrieve.laf?csmSeq='

index = list(range(15))
index.remove(3)
index
```

```python
def autoSoup(url):
    return BeautifulSoup(requests.get(url).text,'lxml')
```

```python
autoSoup(url+'1').findAll('h6','hide')[0].next.next.next.get('href').split('=')[1]
```

```python
nums = []

for i in index:
    temp = autoSoup(url+str(i)).findAll('h6','hide')
    for item in temp:
        nums.append(item.next.next.next.get('href').split('=')[1])
len(nums)
```

```python
nums
```

```python
def clean(text):
    return "".join("".join("".join(text.split('\r')).split('\t')).split('\n'))
```

```python
clean(autoSoup(root_url+nums[0]).findAll('div','ttl')[0].text)
```

```python
clean(autoSoup(root_url+nums[0]).findAll('div','ans')[0].text)
```

```python
final_dict = defaultdict(list)

for i in nums:
    for j in autoSoup(root_url+i).findAll('div','ttl'):
        final_dict['content_q'].append(clean(j.text))
    for j in autoSoup(root_url+i).findAll('div','ans'):
        final_dict['content_a'].append(clean(j.text))
    for j in autoSoup(root_url+i).findAll('li','title'):
        final_dict['major'].append(j.text.split('\n')[1].split(':')[0])
        final_dict['minor'].append(j.text.split('\n')[1].split(':')[1])
    for j in autoSoup(root_url+i).findAll('li','title'):
        final_dict['view_count'].append(j.p.span.text.split('\xa0')[0].strip('건'))
        final_dict['recommend'].append(j.p.span.next.next.next.text.split('\xa0')[0].strip('건'))
```

```python
df = pd.DataFrame(final_dict)
df
```

```python
df['content'] = df['content_q'] + ' / ' + df['content_a']
```
