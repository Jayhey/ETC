```python
import konlpy
import gensim
import pandas as pd
from collections import defaultdict
```

```python
kkma = konlpy.tag.Kkma()
kkma.pos('테스트용 출력입니다. 반갑습니다.')
```

```python
#pos = lambda d: [p for p in kkma.pos(d,True)]
pos = lambda d: ["/".join(p) for p in kkma.pos(d,True) if p[1][0]=='N' or p[1] in ['VV','VA','VXA','VXV']]
```

```python
pos('테스트용 출력입니다. 반갑습니다.')
```

```python
df = pd.read_csv('friendly_law.csv')
df
```

```python
docs_ko
text_ko = []
for i in df.content:
    docs_ko.append(i)
    text_ko.append(pos(i))

text_ko
```

```python
len(text_ko)
```

```python
from gensim import corpora
dictionary_ko = corpora.Dictionary(text_ko)
```

```python
len(dictionary_ko)
```

```python
from gensim import models

tf_ko = []

for text in text_ko:
    tf_ko.append(dictionary_ko.doc2bow(text))

tf_ko
```

```python
tfidf_model_ko = models.TfidfModel(tf_ko)
tfidf_model_ko
```

```python
tfidf_ko = tfidf_model_ko[tf_ko]
```

```python
tfidf_ko[0]
```

```python
user_text= '중고나라에서 물건을 사다가 사기를 당했어요. 돈을 보내줬는데 전화를 안받네요. 어떻게 해야하죠?'

user_pos = pos(user_text)
user_pos
```

```python
user_tf = dictionary_ko.doc2bow(user_pos)
user_tf
```

```python
tfidf_ko[0]
```

```python
tfidf = []
for tf in tfidf_ko:
    temp = {}
    for t in tf:
        temp[t[0]] = t[1]
    tfidf.append(temp.copy())
```

```python
tfidf[0]
```

```python
hubo = defaultdict(int)
for i,tfs in enumerate(tfidf):
    for item in user_tf:
        try:
            hubo[i] += tfs[item[0]]*item[1]
        except:
            pass
        
hubo
```

```python
max(hubo.items(),key=lambda x:x[1])
```

```python
dab = max(hubo.items(),key=lambda x:x[1])[0]
dab
```

```python
docs_ko[dab]
```

```python
def what(user_input):
    user_pos = pos(user_input)
    user_tf = dictionary_ko.doc2bow(user_pos)
    hubo = defaultdict(int)
    for i,tfs in enumerate(tfidf):
        for item in user_tf:
            try:
                hubo[i] += tfs[item[0]]*item[1]
            except:
                pass

    return docs_ko[max(hubo.items(),key=lambda x:x[1])[0]]
```

```python
what('택배기사님이 집 문 앞에 물건을 두고갔는데 사라졌어요. 이거 보상받을 수 있나요?')
```

```python
import time
s_time = time.time()
print(what('택배기사님이 집 문 앞에 물건을 두고갔는데 사라졌어요. 이거 보상받을 수 있나요?'))
e_time = time.time()
print('소요 시간 :%3f'%(e_time-s_time))
```
